# Renewable Energy & Electricity Demand (5-минутное разрешение)

Описанная база данных была создана с использованием данных, полученных от Калифорнийского независимого системного оператора (CAISO) и Национальной лаборатории возобновляемой энергии (NREL). Все данные собирались с интервалом в пять минут, а затем очищались и модифицировались для создания базы данных, состоящей из трех временных рядов: производство солнечной энергии, производство ветровой энергии и спрос на электроэнергию. База данных содержит 12 столбцов, включая:
- дату
- станцию (1: зима, 2: весна, 3: лето, 4: осень)
- день недели (0: понедельник, ... , 6: воскресенье)
- DHI (Вт/м2)
- DNI (Вт/м2)
- GHI (Вт/м2)
- скорость ветра (м/с)
- влажность (%)
- температура (градусы)
- производство солнечной энергии (МВт)
- производство ветровой энергии (МВт)
- спрос на электроэнергию (МВт).
