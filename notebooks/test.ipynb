```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# –ì–µ–Ω–µ—Ä–∏—Ä—É–µ–º —Å–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏–µ –¥–∞–Ω–Ω—ã–µ
np.random.seed(0)
X = np.random.rand(100, 2)  # 2 –ø—Ä–∏–∑–Ω–∞–∫–∞
coef = np.array([3, 5])
y = X @ coef + 2 + np.random.randn(100) * 0.5  # y = 3*x1 + 5*x2 + 2 + —à—É–º

# –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å
model = LinearRegression()
model.fit(X, y)

print("–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã:", model.coef_)
print("–°–º–µ—â–µ–Ω–∏–µ:", model.intercept_)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è - scatter –∏ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –ø–æ —Å–µ—Ç–∫–µ
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X[:,0], X[:,1], y, label='–î–∞–Ω–Ω—ã–µ', alpha=0.5)
x1_grid, x2_grid = np.meshgrid(np.linspace(0,1,20), np.linspace(0,1,20))
y_pred_grid = model.predict(np.c_[x1_grid.ravel(), x2_grid.ravel()]).reshape(x1_grid.shape)

ax.plot_surface(x1_grid, x2_grid, y_pred_grid, color='orange', alpha=0.5)
ax.set_xlabel('–ü—Ä–∏–∑–Ω–∞–∫ 1')
ax.set_ylabel('–ü—Ä–∏–∑–Ω–∞–∫ 2')
ax.set_zlabel('–¶–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫')
plt.title("–ü—Ä–æ—Å—Ç–∞—è –ª–∏–Ω–µ–π–Ω–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è")
plt.legend()
plt.show()
```
