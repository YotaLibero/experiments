```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
y_tr = [3, 5, 6, 7]
y_pr = [3, 2, 6, 6.6]
mse = mean_squared_error(y_tr, y_pr)
```

```python
from IPython.display import Markdown, display

def md(text):
    display(Markdown(text))

md("### üìä –ú–µ—Ç—Ä–∏–∫–∏ –º–æ–¥–µ–ª–∏")
md("| –ú–µ—Ç—Ä–∏–∫–∞ | –ó–Ω–∞—á–µ–Ω–∏–µ |")
md("|---|---|")
md(f"| MAE | **{mse:.2f}** |")
md(f"| RMSE | **{mse:.2f}** |")
```

```python
print("====== –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –º–æ–¥–µ–ª–∏ ======")
print(f"MAE  : {mse:.3f}")
print(f"MSE  : {mse:.3f}")
print(f"RMSE : {mse:.3f}")
```
